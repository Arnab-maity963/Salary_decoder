# Task 2


```python
import pandas as pd
import numpy as np

# Clean column names

df.columns = df.columns.str.strip()

df.columns = (
    df.columns
      .str.lower()
      .str.replace(" ", "_")
)

print("Columns after renaming:")
print(df.columns)

# Standardize Role Names

print("\nUnique Roles Before Cleaning:\n")
print(df["role"].unique())

role_mapping = {

    "SDE Backend": "SDE Backend",
    "Backend Engineer": "SDE Backend",
    "Backend Developer": "SDE Backend",
    "Backend Dev": "SDE Backend",
    "BE": "SDE Backend",

    "SDE Frontend": "SDE Frontend",
    "Frontend Engineer": "SDE Frontend",
    "Frontend Developer": "SDE Frontend",
    "Frontend Dev": "SDE Frontend",
    "FE": "SDE Frontend",

    "SDE Full-Stack": "SDE Full-Stack",
    "Full Stack Engineer": "SDE Full-Stack",
    "Full Stack Developer": "SDE Full-Stack",
    "Full Stack Dev": "SDE Full-Stack",

    "Data Scientist": "Data Scientist",
    "Data Sci": "Data Scientist",
    "DS": "Data Scientist",

    "Data Analyst": "Data Analyst",

    "Business Analyst": "Business Analyst",

    "DevOps": "DevOps Engineer",
    "DevOps Engineer": "DevOps Engineer",

    "UI Designer": "UI/UX Designer",
    "UX Designer": "UI/UX Designer",
    "UI/UX Designer": "UI/UX Designer",

    "Product Manager": "Product Manager"
}

df["role_clean"] = (
    df["role"]
      .astype(str)
      .str.strip()
      .replace(role_mapping)
)


# Standardize Company Type

df["company_type_clean"] = (
    df["company_type"]
      .astype(str)
      .str.strip()
      .str.lower()
)

company_mapping = {

    "mnc":"MNC",
    "unicorn":"Unicorn",
    "mid-size":"Mid-size",
    "mid size":"Mid-size",
    "early-stage":"Early-stage",
    "early stage":"Early-stage"
}

df["company_type_clean"] = df["company_type_clean"].replace(company_mapping)

# Standardize Education Tier

education_mapping = {

    "Tier 1":"Tier 1",
    "Tier-1":"Tier 1",
    "T1":"Tier 1",
    "1":"Tier 1",

    "Tier 2":"Tier 2",
    "Tier-2":"Tier 2",
    "T2":"Tier 2",
    "2":"Tier 2",

    "Tier 3":"Tier 3",
    "Tier-3":"Tier 3",
    "T3":"Tier 3",
    "3":"Tier 3"
}

df["education_tier_clean"] = (
    df["education_tier"]
      .astype(str)
      .str.strip()
      .replace(education_mapping)
)


# Standardize Work Mode

df["work_mode_clean"] = (
    df["work_mode"]
      .astype(str)
      .str.strip()
      .str.title()
)

# Function to Convert CTC into LPA

def parse_ctc(ctc):

    if pd.isna(ctc):
        return np.nan

    ctc = str(ctc).strip()

    ctc = ctc.replace("₹","")
    ctc = ctc.replace("LPA","")
    ctc = ctc.replace("lpa","")
    ctc = ctc.strip()

    if "," in ctc:

        number = ctc.replace(",","")

        try:
            number = float(number)

            if number > 100000:
                return number/100000

            return number

        except:
            return np.nan

    try:

        value = float(ctc)

        return value

    except:

        return np.nan

# Convert Current & Previous CTC

df["current_ctc_lpa"] = df["current_ctc"].apply(parse_ctc)

df["previous_ctc_lpa"] = df["previous_ctc"].apply(parse_ctc)

# Remove Duplicate Records

print("\nDuplicates Before:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicates After:", df.duplicated().sum())

# Handle Missing Values

# Keep Previous CTC as NaN (Freshers)

df["skills"] = df["skills"].fillna("Not Available")

# Verification

print("\n CLEANED DATA TYPES ")
print(df.dtypes)

print("\n ROLE COUNTS ")
print(df["role_clean"].value_counts())

print("\n COMPANY TYPE COUNTS")
print(df["company_type_clean"].value_counts())

print("\n EDUCATION TIER COUNTS ")
print(df["education_tier_clean"].value_counts())

print("\nSAMPLE DATA ")

print(df.head())
```

## Task 2 Summary

- Renamed all columns into snake_case.
- Standardized job roles into canonical names.
- Converted Current_CTC and Previous_CTC into numeric LPA values.
- Standardized Company Type, Education Tier, and Work Mode.
- Removed duplicate records.
- Preserved missing Previous_CTC values for freshers while filling missing Skills with "Not Available".
- The cleaned dataset is now ready for business analysis.

# Task 3

```python

# Q3.1 CTC Distribution by Role

role_salary = (
    df.groupby("role_clean")["current_ctc_lpa"]
      .agg(["median","mean","min","max"])
      .sort_values(by="median", ascending=False)
)

print(role_salary)

highest = role_salary.index[0]
lowest = role_salary.index[-1]

print("\nHighest Paying Role :", highest)
print("Lowest Paying Role :", lowest)
```

### Insight (Q3.1)

The role with the highest median salary is ______.
The lowest paying role is ______.
A large difference between mean and median indicates the presence of salary outliers.

```python

# Q3.2 Experience vs Salary

backend = df[df["role_clean"]=="SDE Backend"].copy()

bins=[-1,1,3,5,100]
labels=["0-1","2-3","4-5","6+"]

backend["exp_band"]=pd.cut(
    backend["years_exp"],
    bins=bins,
    labels=labels
)

exp_salary = backend.groupby("exp_band")["current_ctc_lpa"].median()

print(exp_salary)

growth = exp_salary.pct_change()*100

print("\nGrowth (%)")
print(growth)
```

### Insight (Q3.2)

Median salary increases steadily with experience.
The largest salary jump occurs between ______ and ______ years.

```python

# Q3.3 Skill Premium


sde = df[
    df["role_clean"].isin([
        "SDE Backend",
        "SDE Frontend",
        "SDE Full-Stack"
    ])
]

skills = [
    "AWS",
    "ML",
    "System Design",
    "Kubernetes"
]

for skill in skills:

    with_skill = sde[
        sde["skills"].fillna("").str.contains(skill,case=False)
    ]["current_ctc_lpa"].median()

    without_skill = sde[
        ~sde["skills"].fillna("").str.contains(skill,case=False)
    ]["current_ctc_lpa"].median()

    premium=((with_skill-without_skill)/without_skill)*100

    print(f"{skill}")
    print("With Skill :", round(with_skill,2))
    print("Without Skill :", round(without_skill,2))
    print("Premium :", round(premium,2),"%")
    print("-"*40)
```

### Insight (Q3.3)

The skill with the highest salary premium is ______.
Students should prioritize learning this skill for better compensation.

```python

# Q3.4 Company Premium

backend=df[df["role_clean"]=="SDE Backend"]

company_salary=backend.groupby(
    "company_type_clean"
)["current_ctc_lpa"].median()

print(company_salary)

premium=((company_salary["Unicorn"]-
          company_salary["Mnc"])
          /company_salary["Mnc"])*100

print("\nUnicorn Premium =",round(premium,2),"%")
```

### Insight (Q3.4)

Unicorn companies offer approximately ____% higher median salary than MNCs for the same role.

```python

# Q3.5 Underpaid Professionals

bins=[-1,1,3,5,100]
labels=["0-1","2-3","4-5","6+"]

df["exp_band"]=pd.cut(
    df["years_exp"],
    bins=bins,
    labels=labels
)

group_size=df.groupby([
    "role_clean",
    "company_type_clean",
    "exp_band"
])["employee_id"].transform("count")

df_filtered=df[group_size>=10].copy()

group_median=df_filtered.groupby([
    "role_clean",
    "company_type_clean",
    "exp_band"
])["current_ctc_lpa"].transform("median")

df_filtered["group_median"]=group_median

df_filtered["gap"]=(
    df_filtered["current_ctc_lpa"]-
    df_filtered["group_median"]
)

underpaid=df_filtered.sort_values(
    by="gap"
).head(10)

print(
underpaid[
[
"employee_id",
"role_clean",
"company_type_clean",
"years_exp",
"current_ctc_lpa",
"group_median",
"gap"
]
]
)
```

### Insight (Q3.5)

These professionals earn significantly below the median salary of similar employees and may have strong negotiation opportunities.
